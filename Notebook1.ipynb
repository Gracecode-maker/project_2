```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Let's create our first dataset - Favorite ice cream flavors!
Ice_cream_data = {
    'Flavor' : ['Chocolate', 'Vanila', 'Strawberry', 'Mint', 'Cookie Dough'],
    'Votes' : [25, 36, 24, 15, 30],
    'Colours' : ['Brown', 'Cream', 'Pink', 'Green', 'Belge'],
    'Numbers' : [ 34, 24, 47, 87, 44]
}
```

```python
#bar chart
df1_rename.plot(kind='bar', x="Ice cream flavour", y="Popularity")
plt.title('Ice cream popularity')
plt.xlabel('Ice cream flavour')
plt.ylabel('Popularity')
plt.grid(alpha=0.25)
plt.show()

#line chart
df1_rename.plot(kind='line', x="Ice cream flavour", y="Cost")

plt.title('Ice cream cost')
plt.xlabel('Ice cream flavour')
plt.ylabel('Cost')
plt.grid(alpha=0.25)
plt.show()
```

```python
plt.figure(figsize=(6,6))

df1_rename.set_index("Ice cream flavour")["Popularity"].plot(
    kind='pie',
    autopct='%1.1f%%',
    ylabel=''   # removes the default y-label
)
plt.style.use('ggplot')
plt.title('Ice Cream Popularity - Pie Chart')
plt.show()
```

```python
plt.figure(figsize=(8,5))

# Bar chart
plt.bar(df1_rename["Ice cream flavour"], df1_rename["Popularity"],
        color='red', label='Popularity (Bar)')

# Line chart
plt.plot(df1_rename["Ice cream flavour"], df1_rename["Cost"],
         color='pink', marker='o', linewidth=2, label='Cost (Line)')

plt.title('Ice Cream Cost')
plt.xlabel('Ice cream flavour')
plt.ylabel('Cost')
plt.ylabel('Popularity')
plt.grid(alpha=0.25)
plt.legend()
plt.tight_layout()

plt.show()
```
