```python
# Does the line go up or down? __________Both_____________
# Which year had the highest value? _________No year mentioned in the csv file_________
# What does the line tell you? _________There are a lot of digits mixed up______________
plt.plot(df["sepal.width"], df["sepal.length"])
plt.xlabel("Length")
plt.ylabel("Width")
plt.title("Sepal width and length")
plt.show()
```

```python
# Tallest bar is: __________width of 4.0____________
# Shortest bar is: ________width of 1.5_____________
# This means that: _______There are more sepals with a width of 4.0 that 1.5______________
plt.bar(df["sepal.width"], df["sepal.length"])
plt.xlabel("width")
plt.ylabel("length")
plt.title("Sepal width and length")
plt.show()
```

```python
# Which range has the most values? ________3.0____________
# Are the values mostly big or small? ______big___________
plt.hist(df["sepal.width"], bins=10)
plt.xlabel("Width")
plt.ylabel("Frequency")
plt.title("Histogram")
```
